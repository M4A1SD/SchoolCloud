```python
from ipywidgets import Dropdown, VBox, Output
from IPython.display import display

# Initialize the output widget
output_area = Output()

# Dropdown for makes
make_dropdown = Dropdown(
    options=list(makes),
    description="Make:",
    value=list(makes)[0] if len(makes) > 0 else None
)

# Dropdown for models
modelMark_dropdown = Dropdown(
    options=[],
    description="Model:",
    value=None
)

def on_make_change(change):
    output_area.clear_output()
    with output_area:
        selected_make = change['new']

        # Filter cars by selected make
        cars = data_df[data_df['tozeret_nm'] == selected_make]
        models = cars['kinuy_mishari'].unique()

        # Update model dropdown
        modelMark_dropdown.options = list(models)
        modelMark_dropdown.value = list(models)[0] if len(models) > 0 else None

def on_model_change(change):
    output_area.clear_output()
    with output_area:
        selected_model = change['new']

        # Filter the DataFrame based on the selected model and make
        matching_rows = data_df[
            (data_df['kinuy_mishari'] == selected_model) &
            (data_df['tozeret_nm'] == make_dropdown.value)
        ]

        print(f"Number of unique rows matching both model and make: {len(matching_rows)}")

# Observe dropdown changes
make_dropdown.observe(on_make_change, names='value')
modelMark_dropdown.observe(on_model_change, names='value')

# Display the dropdowns and output area
display(VBox([make_dropdown, modelMark_dropdown, output_area]))
```
